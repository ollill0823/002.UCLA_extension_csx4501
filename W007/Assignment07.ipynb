# Use the model to generate predictions on the test data and then compare against the actual target values by calculating the mean squared error and R-squared values.:

## bmi

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

X=features_df[['bmi']]
y=target_df

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=87)

scaler=StandardScaler()

X_train= scaler.fit_transform(X_train)
X_test= scaler.fit_transform(X_test)
linear_regression=LinearRegression()
model=linear_regression.fit(X_train,y_train)


y_pred= model.predict(X_test)

plt.subplot(131)
plt.scatter(X_train, y_train, color='purple' )
plt.plot(X_train, linear_regression.predict(X_train),color='blue')
plt.xlabel('s5(X_train)',fontsize=14)
plt.ylabel("target(train)",fontsize=14)

plt.subplot(133)
plt.scatter(X_test, y_test, color='coral')
plt.plot(X_test, y_pred, 'b-')
plt.xlabel('s5(X_test)',fontsize=14)
plt.ylabel("target(test)",fontsize=14)
plt.show()

w_0 = linear_regression.intercept_
w_1 = linear_regression.coef_
print('\033[1mbmi test coef1\033[0m:{}'.format(round(w_0,1)))
print('\033[1mbmi test intercpet\033[0m:{}'.format(round(w_1[0],1)))
print('\033[95m\033[1mbmi test equation: y = {} + {}*X({})\033[0m'.format(round(w_0,1), round(w_1[0],1),'bmi'))
print("\033[1mTraining score \033[0m: ", round(linear_regression.score(X_train, y_train),2))
from sklearn.metrics import r2_score, mean_squared_error
r2score = r2_score(y_test, y_pred)
msescore = mean_squared_error(y_test, y_pred)
print("\033[1mTesting score R2 \033[0m: ", round(r2score,2))
print("\033[1mTesting score StdDev \033[0m: ", round(np.sqrt(msescore),1))
print("\033[1mMSE \033[0m: ", round(mean_squared_error(y_pred, y_test),1))
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
best2 = SelectKBest(f_regression, k=2).fit(features_df, target_df)
bestcolumns = features_df.columns[best2.get_support()]
myfeatures=features_df[bestcolumns]


lindf=pd.DataFrame(myfeatures)
reg=LinearRegression().fit(lindf,y)
ypred=reg.predict(lindf)

print('\033[1m{} coef1\033[0m:{}, \033[1m{} coef2\033[0m:{}'.format(myfeatures.columns[0], round(reg.coef_[0],1), 
                                                                 myfeatures.columns[1], round(reg.coef_[1],1)))
print('\033[1mintercpet\033[0m:{}'.format(round(reg.intercept_,1)))
print('\033[95m\033[1mBest2 variables equation y_100% = {} + {}*X({}) + {}*X({})\033[1m'.format(round(reg.intercept_,1), 
                                                            round(reg.coef_[0],1),myfeatures.columns[0], 
                                                              round(reg.coef_[1],1),myfeatures.columns[1]))
ytrain_100=reg.intercept_+reg.coef_*myfeatures

plt.plot(myfeatures.bmi, target_df,'ro', myfeatures.s5, target_df,'yo', myfeatures, ytrain_100, 'b.' )
plt.show()
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
df = pd.DataFrame(X, columns=['x1','x2'])
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn import datasets
import ipywidgets
import plotly.express as px
import pandas as pd
import seaborn as sns
```

# Use sklearn's KMeans to identify 4 clusters

# Generate another scatter plot of the data with the points colored by cluster to see how the algorithm fared

```python

kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
sns.scatterplot(x=df.x1,y=df.x2,hue=kmeans.labels_,palette='rainbow')
```

# Repeat with different parameter values in the call to KMeans, as well as for different parameters in the initial "blob"-formation with make_blobs, to see where KMeans does a good job and where it doesn't

```python
def kmeans_make_blobs(n_clusters):
    kmeans = KMeans(n_clusters)
    kmeans.fit(X)
    sns.scatterplot(x=df.x1,
                    y=df.x2,
                    hue=kmeans.labels_,
                    palette='rainbow')
ipywidgets.interact(kmeans_make_blobs,n_clusters=(1,20));
```
